```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
def test_policy(P, R, policy, test_count=1000, gamma=0.9):
    num_state = P.shape[-1]
    total_episode = num_state * test_count
    # start in each state
    total_reward = 0
    for state in range(num_state):
        state_reward = 0
        for state_episode in range(test_count):
            episode_reward = 0
            disc_rate = 1
            while True:
                # take step
                action = policy[state]
                # get next step using P
                probs = P[action][state]
                candidates = list(range(len(P[action][state])))
                next_state =  choice(candidates, 1, p=probs)[0]
                # get the reward
                reward = R[state][action] * disc_rate
                episode_reward += reward
                # when go back to 0 ended
                disc_rate *= gamma
                if next_state == 0:
                    break
            state_reward += episode_reward
        total_reward += state_reward
    return total_reward / total_episode
```

```python
def trainVI(P, R, discount=0.9, epsilon=[1e-9]):
    vi_df = pd.DataFrame(columns=["Epsilon", "Policy", "Iteration", 
                                  "Time", "Reward", "Value Function"])
    for eps in epsilon:
        vi = ValueIteration(P, R, gamma=discount, epsilon=eps, max_iter=int(1e15))
        vi.run()
        reward = test_policy(P, R, vi.policy)
        info = [float(eps), vi.policy, vi.iter, vi.time, reward, vi.V]
        df_length = len(vi_df)
        vi_df.loc[df_length] = info
    return vi_df
```

```python
import matplotlib.pyplot as plt
import numpy as np
from hiive.mdptoolbox.mdp import ValueIteration, PolicyIteration
from hiive.mdptoolbox.example import forest

def analyze_mdp_solutions_over_gamma(P, R, gamma_values, max_iter=1e6, test_count=1000):
    vi_iters, pi_iters, vi_times, pi_times, vi_policies, pi_policies, vi_rewards, pi_rewards = [], [], [], [], [], [], [], []

    for gamma in gamma_values:
        # Initialize the algorithms with the current gamma value
        vi = ValueIteration(P, R, gamma=gamma, max_iter=max_iter)
        pi = PolicyIteration(P, R, gamma=gamma, max_iter=max_iter)

        # Run the algorithms
        vi.run()
        pi.run()

        # Test the policies and get the average rewards
        vi_avg_reward = test_policy(P, R, vi.policy, test_count=test_count, gamma=gamma)
        pi_avg_reward = test_policy(P, R, pi.policy, test_count=test_count, gamma=gamma)

        # Append the results for plotting and comparison
        vi_iters.append(vi.iter)
        pi_iters.append(pi.iter)
        vi_times.append(vi.time)
        pi_times.append(pi.time)
        vi_policies.append(vi.policy)
        pi_policies.append(pi.policy)
        vi_rewards.append(vi_avg_reward)
        pi_rewards.append(pi_avg_reward)

    return vi_iters, pi_iters, vi_times, pi_times, vi_policies, pi_policies, vi_rewards, pi_rewards

# Define a range of gamma values
gamma_values = np.linspace(0.1, 0.99, 10)

# Get the forest MDP
P, R = forest(S=20, r1=10, r2=6, p=0.1)

# Analyze MDP solutions over different gamma values
vi_iters, pi_iters, vi_times, pi_times, vi_policies, pi_policies, vi_rewards, pi_rewards = analyze_mdp_solutions_over_gamma(P, R, gamma_values, test_count=1000)

# Compare the policies and plot the results
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.plot(gamma_values, vi_iters, label='Value Iteration')
plt.plot(gamma_values, pi_iters, label='Policy Iteration')
plt.xlabel('Gamma')
plt.ylabel('Iterations to Converge')
plt.title('Iterations vs Gamma')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(gamma_values, vi_times, label='Value Iteration')
plt.plot(gamma_values, pi_times, label='Policy Iteration')
plt.xlabel('Gamma')
plt.ylabel('Time to Converge (Seconds)')
plt.title('Convergence Time vs Gamma')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(gamma_values, vi_rewards, label='Value Iteration')
plt.plot(gamma_values, pi_rewards, label='Policy Iteration')
plt.xlabel('Gamma')
plt.ylabel('Average Reward')
plt.title('Reward vs Gamma')
plt.legend()

for i, gamma in enumerate(gamma_values):
    print(f"Gamma: {gamma}")
    print(f"Value Iteration Policy: {vi_policies[i]}")
    print(f"Policy Iteration Policy: {pi_policies[i]}")
    print(f"Are policies equal? {'Yes' if np.array_equal(vi_policies[i], pi_policies[i]) else 'No'}")
    print("-" * 30)

plt.tight_layout()
plt.show()
```

# Q-Learning

```python
def trainQ(P, R, discount=0.9, alpha_dec=[0.99], alpha_min=[0.001], 
           epsilon=[0.9], epsilon_decay=[0.99], n_iter=[1000000]):
    q_df = pd.DataFrame(columns=["Iterations", "Alpha Decay", "Alpha Min", 
                                 "Epsilon", "Epsilon Decay", "Reward",
                                 "Time", "Policy", "Value Function",
                                 "Training Rewards"])
    
    count = 0
    for i in n_iter:
        for eps in epsilon:
            for eps_dec in epsilon_decay:
                for a_dec in alpha_dec:
                    for a_min in alpha_min:
                        q = QLearning(P, R, discount, alpha_decay=a_dec, 
                                      alpha_min=a_min, epsilon=eps, 
                                      epsilon_decay=eps_dec, n_iter=i)
                        q.run()
                        reward = test_policy(P, R, q.policy)
                        count += 1
                        print(f"Run {count}: Iterations={i}, Alpha Decay={a_dec}, Alpha Min={a_min}, "
                              f"Epsilon={eps}, Epsilon Decay={eps_dec}, Time={q.time:.4f} seconds, Reward={reward:.4f}")
                        st = q.run_stats
                        rews = [s['Reward'] for s in st]
                        info = [i, a_dec, a_min, eps, eps_dec, reward, 
                                q.time, q.policy, q.V, rews]
                        
                        df_length = len(q_df)
                        q_df.loc[df_length] = info
    return q_df
```

```python
alpha_decs = [0.99, 0.999]
alpha_mins =[0.001, 0.0001]
eps = [0.8, 0.9, 0.99]
eps_dec = [0.99, 0.999]
iters = [10000, 100000]
q_df = trainQ(P, R, discount=0.9, alpha_dec=alpha_decs, alpha_min=alpha_mins, 
              epsilon=eps, epsilon_decay=eps_dec, n_iter=iters)
```
