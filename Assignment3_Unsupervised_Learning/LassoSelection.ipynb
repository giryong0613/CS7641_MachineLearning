```python
wine_dataset = pd.read_csv("wine_dataset.csv")
ufc_dataset = pd.read_csv("ufc_data.csv")

ufc_dataset.loc[ufc_dataset["Winner"] == "Red", "Winner"] = 1
ufc_dataset.loc[ufc_dataset["Winner"] == "Blue", "Winner"] = 0
ufc_dataset.drop(["no_of_rounds"], axis=1 ,inplace=True)
# al zero
ufc_dataset.drop(["B_draw"], axis=1 ,inplace=True)
ufc_dataset.drop(["R_draw"], axis=1 ,inplace=True)

ufcX = ufc_dataset.drop("Winner", 1).copy().values
ufcY = ufc_dataset["Winner"].copy().values
ufc_x_df = ufc_dataset.drop("Winner",1)

wineX = wine_dataset.drop("class", 1).copy().values
wineY = wine_dataset["class"].copy().values
wine_x_df = wine_dataset.drop("class",1)


min_max_scaler = preprocessing.MinMaxScaler()
ufcX = min_max_scaler.fit_transform(ufcX)
ufcX_df = pd.DataFrame(ufcX, columns=ufc_x_df.columns)
wineX = min_max_scaler.fit_transform(wineX)
wineX_df = pd.DataFrame(wineX, columns=wine_x_df.columns)
```

```python
# Perform feature selection with Lasso, getting 40 most important features
lasso = LassoCV(max_iter=10000).fit(ufcX, ufcY)
importance = np.abs(lasso.coef_)
idx_third = importance.argsort()[-40]
threshold = importance[idx_third] + 1e-5

# Create the SelectFromModel object using the threshold
sfm = SelectFromModel(lasso, threshold=threshold).fit(ufcX, ufcY)

# Transform the dataset to reduce it to the selected features
ufcX_40 = sfm.transform(ufcX)

# Check the number of features selected
n_features = ufcX_40.shape[1]
print(f"UFC Data Set Selected features: {n_features}")

# Get the names of the selected features
selected_feature_names = ufc_dataset.drop("Winner", axis=1).columns[sfm.get_support(indices=True)]
print(f"Selected features: {selected_feature_names.tolist()}")
```

```python
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_squared_error

# Perform LassoCV for feature selection
lasso_errors = {}
lasso = LassoCV(max_iter=10000, cv=5).fit(ufcX, ufcY)  # Using cross-validation
coefs = np.abs(lasso.coef_)

# Sort the coefficients in descending order and their respective indices
sorted_indices = np.argsort(coefs)[::-1]

for num_features in range(1, len(coefs) + 1):
    # Select the top 'num_features' coefficients
    top_indices = sorted_indices[:num_features]

    # Transform X to contain only the features with the top coefficients
    X_top_features = ufcX[:, top_indices]
    
    # Fit the Lasso model again on the reduced dataset (this can be optimized further)
    lasso_top_features = LassoCV(max_iter=10000, cv=5).fit(X_top_features, ufcY)
    
    # Predict using the reduced model to calculate the error
    y_pred = lasso_top_features.predict(X_top_features)
    
    # Calculate RMSE for the current number of features
    rmse = mean_squared_error(ufcY, y_pred, squared=False)
    lasso_errors[num_features] = rmse
    # print(f"Number of features: {num_features}, RMSE: {rmse}")

# Now plot the RMSE against the number of features
plt.figure()
plt.plot(list(lasso_errors.keys()), list(lasso_errors.values()), marker='o')
plt.xlabel("Number of features selected")
plt.ylabel("RMSE")
plt.title("RMSE vs Number of Features Selected by Lasso")
plt.show()
```

```python
# stratified cross validation
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

def stratFold(train_X_all, train_y_all, classifier, metric, average=None,
              fold=5):
    train_metric, val_metric, indices = [], [], []
    for m in range(100, len(train_X_all), 100):
        train_X = train_X_all[:m]
        train_y = train_y_all[:m]
        skfolds = StratifiedKFold(n_splits=fold)
        metric_list = []
        metric_list_train = []
        for train_index, test_index in skfolds.split(train_X, train_y):
            train_X_folds = train_X[train_index, :]
            train_y_folds = train_y[train_index]
            test_X_fold = train_X[test_index, :]
            test_y_fold = train_y[test_index]
            classifier.fit(train_X_folds, train_y_folds)
            train_y_folds_pred = classifier.predict(train_X_folds)
            test_y_fold_pred = classifier.predict(test_X_fold)
            if average:
                metric_list.append(metric(test_y_fold, test_y_fold_pred,
                                     average=average))
                metric_list_train.append(metric(train_y_folds, train_y_folds_pred, 
                                            average=average))
            else:
                metric_list.append(metric(test_y_fold, test_y_fold_pred))
                metric_list_train.append(metric(train_y_folds, train_y_folds_pred))
                
        test_avg = sum(metric_list)/len(metric_list)
        train_avg = sum(metric_list_train)/len(metric_list_train)
        val_metric.append(test_avg)
        train_metric.append(train_avg)
        indices.append(m)
    
    return train_metric, val_metric, indices
```

```python
def plot_complexity(train_scores, val_scores, indices, title, inverse_x=True):
    plt.plot(indices, train_scores , "r-+", linewidth=2, label="train")
    plt.plot(indices, val_scores, "b-+", linewidth=2, label="validation")
    plt.title(title)
    plt.legend()
    plt.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.3)
    if inverse_x:
        plt.gca().invert_xaxis()
```

```python
def plot_learning_curve(train_scores, val_scores, indices, title, inverse_x=False):
    plt.plot(indices, train_scores , "r-+", linewidth=2, label="train")
    plt.plot(indices, val_scores, "b-+", linewidth=2, label="validation")
    plt.title(title)
    plt.legend()
    plt.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.3)
    if inverse_x:
        plt.gca().invert_xaxis()
```
