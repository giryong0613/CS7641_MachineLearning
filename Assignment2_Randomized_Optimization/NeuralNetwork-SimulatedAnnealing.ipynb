## Neural Network - Simulated Anealing

```python
import sklearn
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.model_selection import train_test_split
pd.options.display.max_columns = None
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('figure', figsize=[12,6])
import mlrose_hiive
from sklearn.metrics import f1_score
from functools import partial
```

```python
np.random.seed(1)
wine_dataset = pd.read_csv("wine_dataset.csv")
print(wine_dataset.info())
wine_dataset.head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Separating the target and the rest of the dataset
target = wine_dataset["class"]
wine_dataset.drop(columns=["class"], inplace=True)

# Scaling the data using MinMaxScaler
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(wine_dataset)

# Transforming the scaled data back to a DataFrame
wine_dataset = pd.DataFrame(x_scaled, columns=wine_dataset.columns)

# Adding the target back to the dataset at the beginning
wine_dataset.insert(0, "class", target)

# Displaying the first few rows
wine_dataset.head(5)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Set up stratified sampling
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=30)

for train_ind, test_ind in split.split(wine_dataset, wine_dataset["class"]):
    train_set = wine_dataset.loc[train_ind]
    test_set = wine_dataset.loc[test_ind]

# Display dataset information
train_set.info()
test_set.info()

# Create X and Y for train and test set
train_X = train_set.drop("class", axis=1)
train_y = train_set["class"]
test_X = test_set.drop("class", axis=1)
test_y = test_set["class"]
```

```python
%%time

# Define an exponential decay schedule for simulated annealing.
from mlrose_hiive import ExpDecay

f1_labs = partial(f1_score, average="weighted")

# Set the hyperparameters for grid search with simulated annealing.
grid_search_parameters = ({
  'schedule': [ExpDecay(10), ExpDecay(50), ExpDecay(100),ExpDecay(1000), ExpDecay(5000)],
  'learning_rate': [0.01, 0.001, 0.0025],  
  'activation': [mlrose_hiive.neural.activation.relu],
  'max_iters': [10000]
})

# Initialize the NNGSRunner to perform grid search for neural network optimization using SA.
nnr = mlrose_hiive.NNGSRunner(
    x_train=train_X,
    y_train=pd.get_dummies(train_y.values.ravel()).values,
    x_test=test_X,
    y_test=pd.get_dummies(test_y.values.ravel()).values,
    experiment_name='nn_test',
    seed=10,
    output_directory="./sa",
    hidden_layer_sizes=[[60,60]],                             
    algorithm=mlrose_hiive.algorithms.sa.simulated_annealing,
    grid_search_parameters=grid_search_parameters,
    grid_search_scorer_method=f1_labs,
    iteration_list=[10000],
    n_jobs=-2
)

# Execute grid search to find the best hyperparameters using simulated annealing.
results = nnr.run() 
```

```python
print(pd.DataFrame(results[0]).columns)
print(pd.DataFrame(results[1]).columns)
print(pd.DataFrame(results[2]).columns)
```

```python
# Convert the result to a dataframe
results_df = pd.DataFrame(results[0])

# Displaying the parameters and their respective fitness/performance
for index, row in results_df.iterrows():
    print("Parameters and Performance:")
    print(f"  Learning Rate: {row['learning_rate']}")
    print(f"  Schedule Type: {row['schedule_type']}")
    print(f"  Initial Temperature: {row['schedule_init_temp']}")
    print(f"  Exponential Decay Constant: {row['schedule_exp_const']}")
    print(f"  Minimum Temperature: {row['schedule_min_temp']}")
    print(f"  Current Value of Schedule: {row['schedule_current_value']}")
    print(f"  Activation: {row['activation']}")
    print(f"  Fitness (Error/Loss value): {row['Fitness']}")
    print(f"  Iteration: {row['Iteration']}")
    print(f"  Evaluation Function Calls (FEvals): {row['FEvals']}")
    print(f"  Time Taken: {row['Time']} seconds")
    print("---------------------------")

# Plotting the performance:
import matplotlib.pyplot as plt

# Plot Fitness over Iterations
plt.figure(figsize=(10,6))
plt.plot(results_df['Iteration'], results_df['Fitness'], marker='o')
plt.title('Fitness over Iterations')
plt.xlabel('Iterations')
plt.ylabel('Fitness (Error/Loss Value)')
plt.grid(True)
plt.show()
```

```python
# Convert the result to a dataframe
results_df = pd.DataFrame(results[0])

# Find the row with the best (minimum) fitness
best_row = results_df.loc[results_df['Fitness'].idxmin()]

print("Best Parameters and Performance:")
print(f"  Learning Rate: {best_row['learning_rate']}")
print(f"  Schedule Type: {best_row['schedule_type']}")
print(f"  Initial Temperature: {best_row['schedule_init_temp']}")
print(f"  Exponential Decay Constant: {best_row['schedule_exp_const']}")
print(f"  Minimum Temperature: {best_row['schedule_min_temp']}")
print(f"  Current Value of Schedule: {best_row['schedule_current_value']}")
print(f"  Activation: {best_row['activation']}")
print(f"  Best Fitness (Error/Loss value): {best_row['Fitness']}")
print(f"  Iteration: {best_row['Iteration']}")
print(f"  Evaluation Function Calls (FEvals): {best_row['FEvals']}")
print(f"  Time Taken: {best_row['Time']} seconds")
```

```python
# 1. Convert the results to a DataFrame.
results_df = pd.DataFrame(results[0])

# 2. Identify the best parameter set based on the minimum fitness.
best_row = results_df.loc[results_df['Fitness'].idxmin()]

# 3. Filter the results for the best parameter set.
best_params_data = results_df[
    (results_df['schedule'] == best_row['schedule']) &
    (results_df['learning_rate'] == best_row['learning_rate']) &
    (results_df['activation'] == best_row['activation']) &
    (results_df['max_iters'] == best_row['max_iters'])
]

# 4. Compute the metrics based on the filtered data.
print(f"Simulated Annealing Neural Network - Results for Best Parameters (Schedule: {best_row['schedule']}, Learning Rate: {best_row['learning_rate']}, Activation: {best_row['activation']}, Max Iters: {best_row['max_iters']}):")
print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
print("Average Fitness:", best_params_data.Fitness.mean())
print("Maximum Fitness:", best_params_data.Fitness.max())
print("Average Time for an Iteration:", best_params_data.Time.mean())
print("\n")
```

```python
# Assuming the results contains curve data for SA
sa_run_curves = pd.DataFrame(results[1])

# Find the parameters corresponding to the run with the highest fitness value for SA
best_params_sa = sa_run_curves.iloc[sa_run_curves.Fitness.idxmax()]

# Get parameters corresponding to the best run
best_schedule_init_temp_sa = best_params_sa['schedule'].init_temp
best_learning_rate_sa = best_params_sa['learning_rate']

# Filter the curves to only include data from the best SA run
best_sa_curve = sa_run_curves[(sa_run_curves['schedule'].apply(lambda s: s.init_temp) == best_schedule_init_temp_sa) & 
                              (sa_run_curves['learning_rate'] == best_learning_rate_sa)]

# Plotting the best SA curve's fitness over iterations
ax_sa = best_sa_curve.Fitness.plot(lw=2, 
                                   color='blue',   # Set the line color to green for distinction
                                   marker='.', 
                                   markersize=2, 
                                   figsize=(12,8), 
                                   grid=True,      # Enable grid
                                   title='Best SA')

# Set x-axis label with increased font size
ax_sa.set_xlabel("Iterations", fontsize=14)

# Set y-axis label with increased font size
ax_sa.set_ylabel("Value", fontsize=14)

# Set the title's font size larger
ax_sa.set_title('Best SA', fontsize=16)

# Set grid appearance to grey and dotted
ax_sa.grid(True, linestyle='--', which='both', color='grey', alpha=.7)
```

```python
# Importing required metrics
from sklearn.metrics import classification_report

# Predicting the test set results using the trained model
y_pred = results[3].predict(test_X)

# Converting ground truth labels to one-hot encoded format
y_true_one_hot = pd.get_dummies(test_y.values.ravel()).values

# Printing the classification report
print(classification_report(y_true_one_hot, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_true_one_hot.argmax(axis=1), y_pred.argmax(axis=1))
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
